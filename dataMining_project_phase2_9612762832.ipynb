```python
'''THIS PREPROCESSING IS DOWN FOR THE THIRD SECTION'''
divar['created_at'] = divar['created_at'].str.split().str[0]

'''THIS PREPROCESSING IS DOWN FOR THE FOURTH SECTION'''
divar["cat1"] = divar["cat1"].astype(str)
divar["cat2"] = divar["cat2"].astype(str)
divar["cat3"] = divar["cat3"].astype(str)
#divar["cat3"] = le.fit_transform(divar["cat3"].astype(str))
divar[['cat1', 'cat2','cat3']] = np.sort(divar[['cat1', 'cat2','cat3']], axis=1)

divar["item_id"] = divar.groupby(by=['cat1', 'cat2','cat3']).ngroup().astype(str) #add item_id column
```

# PART 1

```python
cities = divar.groupby(by=['city'])
transaction_divar = []
for city, grp in cities:
    #transaction_divar.append(tuple(items['cat1']))
    itm_lst = tuple(grp['cat1'])
    if len(itm_lst)>0:
      transaction_divar.append(itm_lst)
    '''transaction_divar will be a list where each of the cells represent one of the cities
    and they are the itmes which were in the transactions of that city'''

```

```python
itemsets1, rules1 = apriori(transaction_divar, min_support=0.5, min_confidence=0.75) 
print(rules1)
```

# PART 2

```python
'''the reason we don't use ID_order is that in this dataset each row represents one
 order which has only one item (so no rules will be found)'''
cust_ids = digikala_orders.groupby(by=['ID_Customer'])
transaction_digi = []
for id, grp in cust_ids:
    #transaction_divar.append(tuple(items['cat1']))
    itm_lst = tuple(grp['ID_Item'].values)
    if len(itm_lst)>0:
      transaction_digi.append(itm_lst)
```

```python
print(len(transaction_digi)) #we have 28700 transactions here 
pattern = pyfpgrowth.find_frequent_patterns(transaction_digi, 3) #min_support is 3
print(pattern)
```

```python
rules = pyfpgrowth.generate_association_rules(pattern, 0.2) #min_conf is 0.3
print(rules)
```

# PART 3

```python

transaction_divar_time = []
for index,items in divar.iterrows():
    #transaction_divar.append(tuple(items['cat1']))
    itm_lst = [items['created_at'],items['platform']]
    if '' not in itm_lst:
      transaction_divar_time.append(itm_lst)

```

```python
itemsets3, rules3 = apriori(transaction_divar_time, min_support=0.1, min_confidence=0.92) 
```

```python
print(rules3)
```
